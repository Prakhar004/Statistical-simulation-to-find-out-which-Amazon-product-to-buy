# Confused b/w which product to buy on Amazon?
## One with higher rating but less no of people who rated it or the one with relatively lower rating but more no of people who rated it?

I created this model by considering 100 people with an average of 4.1 star rating on one product and 1000 people with an average rating of 4.0 star on another product.
You are free to play around with the parameter n1, n2 and their fraction of rating in each star.

```python
#Lets start with Importing necessary libraries

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def get_avg_rating(rating_distr, n):
    vals=np.random.random(n).reshape(-1,1)
    ratings= 1 + np.argmax(vals < rating_distr.cumsum(), axis=1)
    return ratings.mean()

#This is a Monte Carlo Simulation, more about it here: https://www.investopedia.com/terms/m/montecarlosimulation.asp
```

### This is our product 1 rating, a small sample ratings, n1 specifies the no of people who rated that product on Amazon or any shopping page

```python
rating_distr_1= np.array([.05,.05,.05,.45,.4]) 
#This should add to 1, as this specifies the fraction of rating for each star (1,2,3,4,5)
n1=100
small_samp_means= np.array([get_avg_rating(rating_distr_1, n1) for sampling in range(10000)])
```

## Plotting the distribution of rating of product 1

```python
plt.figure(figsize=(10,4))
plt.bar([str(i) + ' star' for i in np.arange(1,6)], rating_distr_1)
plt.xticks(np.arange(5), fontsize= 20)
plt.yticks(np.arange(0,.5,.1), fontsize=14)
plt.ylabel('Fraction', fontsize=20)
plt.title('n=%s'%n1, fontsize=30)

#Plotting to see the probability distribution 
                                                
```

```python
plt.figure(figsize=(10,4))
sns.distplot(small_samp_means, bins=20)
plt.xlabel('Avg Rating', fontsize=20)
plt.ylabel('Density', fontsize=20)
plt.xticks(np.arange(0,5.1,.1), fontsize=14)
mu,dev = small_samp_means.mean(), small_samp_means.std()
plt.xlim(mu-4*dev, mu+4*dev)
plt.title('Avg=%s\nDev=%s'%(mu,dev), fontsize=20)
plt.axvline(mu+dev, color='r', linestyle='--')
plt.axvline(mu-dev, color='r', linestyle='--')
```

### This is our product 2 rating, a large sample ratings, n2 specifies the no of people who rated that product on Amazon or on any other shopping platform

```python
rating_distr_2= np.array([.05,.05,.05,.55,.3]) 
#this should add to 1, as this specifies the fraction of rating in each star
n2=1000
large_samp_means= np.array([get_avg_rating(rating_distr_2, n2) for sampling in range(10000)])
```

### Plotting the distribution of rating of product 2

```python
plt.figure(figsize=(10,4))
plt.bar([str(i) + ' star' for i in np.arange(1,6)], rating_distr_2)
plt.xticks(np.arange(5), fontsize= 20)
plt.yticks(np.arange(0,.5,.1), fontsize=14)
plt.ylabel('Fraction', fontsize=20)
plt.title('n=%s'%n2, fontsize=30)

```

```python
plt.figure(figsize=(10,4))
sns.distplot(large_samp_means, bins=20)
plt.xlabel('Avg Rating', fontsize=20)
plt.ylabel('Density', fontsize=20)
plt.xticks(np.arange(0,5.1,.1), fontsize=14)
mu,dev = large_samp_means.mean(), large_samp_means.std()
plt.xlim(mu-4*dev, mu+4*dev)
plt.title('Avg=%s\nDev=%s'%(mu,dev), fontsize=20)
plt.axvline(mu+dev, color='r', linestyle='--', alpha=0.5)
plt.axvline(mu-dev, color='r', linestyle='--', alpha=0.5)
```

```python
plt.figure(figsize=(10,4))
sns.distplot(small_samp_means, bins=20)
sns.distplot(large_samp_means, bins=20)
plt.xlabel('Avg_Rating', fontsize=20)
plt.ylabel('Density', fontsize=20)

mu, sigma= small_samp_means.mean(), small_samp_means.std()
#plt.xlim(mu-4*dev,mu+4*dev)
plt.legend(['n=%s'%n1, 'n=%s'%n2], fontsize=20)
plt.axvline(small_samp_means.mean(), color ='k', linestyle='--')
plt.axvline(large_samp_means.mean(), color ='k', linestyle='--')
```

```python
prob_= (small_samp_means>large_samp_means).mean()

plt.figure(figsize=(10,4))
sns.distplot(small_samp_means- large_samp_means, bins=20)
plt.xlabel('Small Samp Avg - Large Samp Avg', fontsize=20)
plt.axvline(0, color='k', linestyle='--')
plt.title('P(small samp  avg > large samp avg) = %s'%round(prob_,2), fontsize=20)
```

```python
print("Probability that a small sample average rating product will be above the large sample average rating is %f percent"%(round(prob_,2)*100))
```

### Assumptions I made while making this?

I took this ditribution from these say x number of people from Amazon and treated it as the population distribution.
But as the sample size increases you have more confidence in the distribution that you are seeing is very close to the population distribution.

### How can I improve this simulation a little more?

1) Firstly the reviews are collected as it is from Amazon site, some reviews for example are given bad if the product packaging wasn't good, or the courier service wasnt on time, which clearly doesn't indicate the product rating but the faults from the seller or courier partner side. So there is a room for removing those reviews, using web scraping and building a simple NLP model that can segregate the reviews as legit or non legit.
##### 
2) If the same product is listed on multiple shopping platforms, we can collect the reviews from all those sites to get rid of the assumption we made above. But it isnt that easy, it may involve a bias towards how the platforms handles the reviews on their platforms. As some platforms add some reviews already by themselves to increase/initiate the sell of that product. Or some platforms doesnt let bad reviews grow on a product more than a certain limit.
